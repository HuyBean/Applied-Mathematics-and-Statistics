```python
from rembg import remove
import numpy as np
from PIL import Image, ImageFilter, ImageDraw 
import matplotlib.pyplot as plt
import time 
```

```python
input_ = 'DSCF9441_Frame.png'
output_  = 'removed.png'

intput_image = Image.open(input_)
output = remove(intput_image)

# Hin th nh gc v nh sau khi gim s lng mu
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].set_title("Original Image")
axs[0].imshow(intput_image)
axs[0].axis('off')
axs[1].set_title("Removed background Image")
axs[1].imshow(output)
axs[1].axis('off')
plt.show()
output.save(output_)
```
